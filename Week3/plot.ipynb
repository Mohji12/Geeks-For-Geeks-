# Import Libraries

```python
import matplotlib.pyplot as plt

import seaborn as sns 

import pandas as pd 

import numpy as np
```

# Data

```python
x=[1,3,5]
y=[2,5,9]
```

```python
sns.set_style("whitegrid")
#sns.set(rc={'axes.facecolor':'cornflowerblue', 'figure.facecolor':'cornflowerblue'})
```

```python
plt.figure(figsize=(20,6),facecolor="lightgreen")
plt.scatter(x,y,color="red",s=100,marker="^")
plt.plot(x,y,linewidth=2,color="black",linestyle="-.")
plt.xlabel("Demand",size=20,color="blue",fontweight='bold')
plt.ylabel("Price",size=20,color="red",fontweight='bold')
plt.title("Our First Plot In Python",size=20)

```

```python
# Plot for sinx 

x=range(1,1000)
x=np.array(x)/20
y=np.sin(x)
```

```python
plt.figure(figsize=(20,6),facecolor="lightgreen")
plt.scatter(x,y,color="red",s=100,marker="^")
plt.plot(x,y,linewidth=2,color="black",linestyle="-.")
plt.xlabel("X Values",size=20,color="blue",fontweight='bold')
plt.ylabel("Sin(X) ",size=20,color="red",fontweight='bold')
plt.title("Sinx ",size=20)
plt.show()
```

```python
x=range(1,1000)
x=np.array(x)/20
y1=np.sin(x)
y2=np.cos(x)


plt.figure(figsize=(20,6),facecolor="lightgreen")
plt.scatter(x,y1,color="red",s=100,marker="^")
plt.plot(x,y1,linewidth=2,color="black",linestyle="-.")

plt.scatter(x,y2,color="green",s=100,marker=">")
plt.plot(x,y2,linewidth=2,color="black",linestyle="-.")

plt.xlabel("X Values",size=20,color="blue",fontweight='bold')


plt.title("sinx and cosx ",size=20)

```

```python
x=range(-150,150)
x=np.array(x)/50
y1=x**2
y2=x**3
y3=x**4


plt.figure(figsize=(20,6),facecolor="lightgreen")
plt.scatter(x,y1,color="red",s=20,marker="^",label="X^2")
plt.plot(x,y1,linewidth=2,color="black",linestyle="-.")

plt.scatter(x,y2,color="green",s=20,marker=">",label="X^3")
plt.plot(x,y2,linewidth=2,color="black",linestyle="-.")


plt.scatter(x,y3,color="blue",s=20,marker=".",label="X^4")
plt.plot(x,y3,linewidth=2,color="black",linestyle="-.")


plt.xlabel("X Values",size=20,color="blue",fontweight='bold')


plt.title("polynomial",size=20)
plt.legend(loc="best",)
```
