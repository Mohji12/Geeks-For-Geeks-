#### Introduction to Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

# Pandas is used for processing data like data cleaning,data wrangling,data manupulation
```

```python
import seaborn as sns
sns.set_palette('magma')
data = sns.load_dataset("iris")
print(data.head())
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
# Basic Operation in pandas
# df.head() # top 5 rows of the data
# df.tail() # last five rows
# df.sample() # to get random data
# df.info() # to get the info of the data
# df.describe() # to get the summary of the data
# df.columns # to get the column names
# df.index # to get the index of the data
# df.shape # to get the number of rows and columns
# df.size # to get the total number of cells in the data
# df.isnull().sum() # to get the number of missing values in the data
# df['column_name'].value_counts() # to get the count of unique values in a column
# df['column_name'].unique() # to get the unique values in a column
# df['column_name'].nunique() # to get the number of unique values in a column
# df['column_name'].dtype # to get the data type of a column
# df['column_name'].mean() # to get the mean of a column
# df['column_name'].median() # to get the median of a column
# df['column_name'].std() # to get the standard deviation of a column
# df['column_name'].max() # to get the maximum value of a column
# df['column_name'].min() # to get the minimum value of a column
# df['column_name'].quantile() # to get the quantile of a column
# df['column_name'].corr() # to get the correlation of a column
# df['column_name'].skew() # to get the skewness of a column
# df['column_name'].kurt() # to get the kurtosis of a column
# df['column_name'].value_counts().plot(kind='bar') # to plot a bar chart
# df['column_name'].value_counts().plot(kind='pie') # to plot a pie chart
# df['column_name'].hist() # to plot a histogram
# df['column_name'].plot(kind='line') # to plot a line chart
# df['column_name'].plot(kind='scatter') # to plot a scatter chart
# df['column_name'].plot(kind='box') # to plot a box plot
# df['column_name'].plot(kind='violin') # to plot a violin plot
# df['column_name'].plot(kind='kde') # to plot a kernel density estimation plot
# df['column_name'].plot(kind='area') # to plot an area chart
# df['column_name'].plot(kind='barh') # to plot a horizontal bar chart
```

```python
print(df['sepal_length'].mean())
print(df['sepal_length'].median())
print(df['sepal_length'].mode())
print(df['sepal_length'].min())
print(df['sepal_length'].max())
print(df['sepal_length'].std())
print(df['sepal_length'].var())
print(df['sepal_length'].quantile(0.25))
print(df['sepal_length'].quantile(0.50))
print(df['sepal_length'].quantile(0.75))
```

```python
# loc and iloc function
# while choosing single column we use single or double brackets and for multiple column we use double square bracket

df[['petal_length']]
df[['petal_length','petal_width']]
```

```python
# loc function is used while accessing the data based column name
# loc work on the names of rows and Columns
df.loc[0:4,"sepal_length":"sepal_width"]
# random rowa and column we will use comman insted of colon
df.loc[[0,12,13],['sepal_length','species']]
```

```python
# iloc work based on index of Rows and Columns
df.iloc[[2,4,6],[0,4]]
```

```python
# Filtering in pandas

df[(df['petal_length'] > 2) & (df['petal_width'] < 2)] # AND
df[(df['petal_length'] > 2) | (df['petal_length'] < 5)] # OR

df[df['species'] == "versicolor"]
```

```python
print(df[df['species'] == 'versicolor']['petal_length'].mean())
print(df[df['species'] == 'setosa']['petal_length'].mean())
print(df[df['species'] == 'virginica']['petal_length'].mean())
```

```python
df.groupby(['species'])['petal_width'].mean()
```

```python
# Sorting the Data

df.sort_values(["sepal_length"],ascending=False)
```

```python
# Changing the Column Names

df['PETAL_LENGTH'] = df['petal_length']
```

```python
df = df.drop(columns="PETAL_LENGTH")
```
