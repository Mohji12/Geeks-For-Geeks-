```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate dummy data for 100 rows
num_rows = 100

# Independent variables
cat_var1 = np.random.choice(['A', 'B', 'C'], size=num_rows)  # Categorical variable 1
cat_var2 = np.random.choice(['X', 'Y', 'Z'], size=num_rows)  # Categorical variable 2
cont_var = np.random.randn(num_rows)  # Continuous variable

# Dependent variable (random continuous data)
dependent_var = np.random.randn(num_rows)

# Create a DataFrame
dummy_data = pd.DataFrame({
    'Categorical_Var1': cat_var1,
    'Categorical_Var2': cat_var2,
    'Continuous_Var': cont_var,
    'Dependent_Var': dependent_var
})

# Display the first few rows
print(dummy_data.head())

```

```python
dummy_data.describe()
```

```python
dummy_data['Categorical_Var1'].value_counts().plot(kind='pie',autopct = '%0.2f')
plt.title('Distribution of Categorical_Var1')
plt.show()
dummy_data['Categorical_Var2'].value_counts().plot(kind='pie',autopct = '%0.2f')
plt.title('Distribution of Categorical_Var2')
plt.show()
```

```python
categorical_data = dummy_data.select_dtypes(include=['object']).columns.tolist()
categorical_data
```

```python
# Split categorical and numerical columns
categorical_columns = dummy_data.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_columns = dummy_data.select_dtypes(include=[np.number]).columns.tolist()

print("Categorical Columns:", categorical_columns)
print("Numerical Columns:", numerical_columns)
```

```python
# Separate DataFrames for categorical and numerical columns
categorical_df = dummy_data[categorical_columns]
numerical_df = dummy_data[numerical_columns]

print("Categorical DataFrame:")
print(categorical_df.head())

print("\nNumerical DataFrame:")
print(numerical_df.head())
```

```python
for i in categorical_df.columns:
    print(i)
    print(categorical_df[i].unique())  
    categorical_df[i].value_counts().plot(kind='pie',autopct="%0.2f")
    plt.show()
```

```python
# Creating a plot of Number,NumberSquare and NumberCuberoot
number = []
for i in range(1,101):
    number.append(i)

number_sqr = []
for i in range(1,101):
    number_sqr.append(i**2)

number_cube = []
for i in range(1,101):
    number_cube.append(i**3)
```

```python
print(number)
print(number_sqr)
print(number_cube)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

# Plot the 3d line
plt.plot(number,number_sqr,number_cube,label = '3D plot',color = 'purple',linewidth = 4)
ax.set_xlabel('x Label')
ax.set_ylabel('y_label')
ax.set_zlabel('Z Label')
plt.legend()

plt.show()
```

```python
# Create a 3D line plot
import plotly.express as go1
import plotly.graph_objects as go
fig = go.Figure(data=[go.Scatter3d(x=number, y=number_sqr, z=number_cube, mode='lines', line=dict(color='blue', width=4))])

# Set axis labels
fig.update_layout(scene=dict(
                    xaxis_title='X Axis',
                    yaxis_title='Y Axis',
                    zaxis_title='Z Axis'),
                  title='3D Line Plot')

# Show the plot
fig.show()
```
